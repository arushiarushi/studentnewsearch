```python
import pyterrier as pt
if not pt.started():
  pt.init()
```

```python
files = pt.io.find_files("/content/drive/MyDrive/Data/www.dailyuw.com")
indexer = pt.FilesIndexer("/content/drive/MyDrive/Data/dailyuw_index",
                          meta={"docno":20,"filename":1024,"title":1024},meta_tags={"title":"title"})
indexref = indexer.index(files)
index = pt.IndexFactory.of(indexref)
print (index.getCollectionStatistics().toString())
```

```python
import pandas as pd
```

```python
queries = pd.DataFrame([["q1","huskies"], ["q2","washington"], ["q3","quarter"], ["q4","Ave"]], columns=["qid","query"])
```

```python
index = pt.IndexFactory.of("/content/drive/MyDrive/Data/dailyuw_index(1)/data.properties")
tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
tf_idf.transform(queries)
```
